```python
import numpy as np
import matplotlib.pyplot as plt  # Fixed the import statement

# Generate a simple data set
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # Fixed the random number generation
y = (4 + 3 * X + np.random.randn(100, 1) > 0).astype(int)  # Fixed the syntax and added missing parenthesis

def sigmoid(z):  # Added a colon at the end of the function definition
    return 1 / (1 + np.exp(-z))

# Initialize weight
theta = np.random.randn(2, 1)
x_b = np.c_[np.ones((100, 1)), X]  # Fixed the syntax for concatenating arrays

# Learning rate
learning_rate = 0.01

# Number of iterations
n_iterations = 1000

# Gradient descent without explicit epochs
for _ in range(n_iterations):  # Fixed the syntax
    # Calculate predicted probabilities
    y_proba = sigmoid(x_b.dot(theta))

    # Update weights using gradients
    gradients = x_b.T.dot(y_proba - y)
    theta = theta - learning_rate * gradients

# Plot the dataset and decision boundaries
plt.scatter(X, y)
plt.xlabel("X")  # Fixed the function name
plt.ylabel("y")  # Fixed the function name

# Plot the decision boundaries
x_decision_boundaries = np.linspace(np.min(X), np.max(X), 100)
x_decision_boundaries = np.c_[np.ones((100, 1)), x_decision_boundaries]
y_decision_boundaries = sigmoid(x_decision_boundaries.dot(theta))
plt.plot(x_decision_boundaries[:, 1], y_decision_boundaries, "-r", label="decision boundaries")  # Updated the plotting arguments
plt.legend()
```
